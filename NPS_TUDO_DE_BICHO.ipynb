```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Respostas NPS - Teste CX.xlsx', index_col='Cliente')
```

### Antes de partir para a criação da visualização, é necessário verificar o *dataset*

```python
df.head()
```

```python
all_words = ' '.join(s for s in df['Observações'])
print('Quantidade de palavras: {}'.format(len(all_words)))
```

### Visualização do WordCloud com os dados qualitativos da NPS

```python
#lista de stopword

stopwords = set(STOPWORDS)

#gerar uma wordcloud

wordcloud = WordCloud(stopwords=stopwords,
                      background_color='white').generate(all_words)

#mostrar imagem final

fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')

plt.tight_layout()
```

### Veja como algumas palavras como "de", "na", "que", "foi", "da" poluem a visualização. É necessário retirá-las.

```python
#lista de stopword

stopwords = set(STOPWORDS)
stopwords.update(["de", "na", "que", "foi", "da"])
#gerar uma wordcloud

wordcloud = WordCloud(stopwords=stopwords,
                      background_color='white', 
                      width=1600, height=800).generate(all_words)

#mostrar imagem final

fig, ax = plt.subplots(figsize=(16,8))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
plt.tight_layout()
```

### Baixar a imagem

```python
#baixando o wordcloud

wordcloud.to_file("nps.png",)
```
